```python
#!/usr/bin/env python
# coding: utf-8

# In[16]:


# import tkinter module
from tkinter import *
 
# make a window
window = Tk()
 
# specify it's size
window.geometry("800x700")
 

# set the background colour of GUI window
window.configure(background="light grey")
 
# set the title of GUI window
window.title("Cafe PoS")
    
# label it in the background
label17 = Label(window)
# position the image as well
label17.place(x=0, y=0)
 
    
# function to calculate the price of all the orders
def calculate():
 
    # dictionary for storing the food quantity and price
    dic = {'fried rice': [e1, 300],
           'noodles': [e2, 200],
           'biryani': [e3, 300],
           'chicken wings': [e4, 200],
           'tea': [e5, 30]}
    total = 0
    for key, val in dic.items():
        if val[0].get() != "":
            total += int(val[0].get())*val[1]
 
    label_total_bill = Label(window,
                    text="Your Total Bill is - "+str(total),
                    font="times 18")
 
    # position it
    label_total_bill.place(x=20, y=490)
    label_total_bill.after(1000, label_total_bill.destroy)
    window.after(1000, calculate)
 
 
label8 = Label(window,
               text="CAFE",
               font="Roman 28 bold")
label8.place(x=350, y=20, anchor="center")
 
label1 = Label(window,
               text="MENU",
               font="times 28 bold")
 
label1.place(x=520, y=70)
 
label2 = Label(window, text="fried rice  Rs 300", font="times 18")
 
label2.place(x=450, y=120)
 
label3 = Label(window, text="noodles Rs 200", font="times 18")
 
label3.place(x=450, y=150)
 
label4 = Label(window, text="biryani  Rs 300", font="times 18")
label4.place(x=450, y=180)
 
label5 = Label(window, text="chicken wings Rs 200", font="times 18")
label5.place(x=450, y=210)
 
label6 = Label(window, text="tea Rs 30", font="times 18")
label6.place(x=450, y=240)
 
# billing section
label9 = Label(window, text="Item order", font="times 18")
label9.place(x=115, y=70)
 
label10 = Label(window,text="fried rice",font="times 18")
label10.place(x=20, y=120)
 
e1 = Entry(window)
e1.place(x=20, y=150)
 
label11 = Label(window, text="noodles",font="times 18")
label11.place(x=20, y=200)
 
e2 = Entry(window)
e2.place(x=20, y=230)
 
label12 = Label(window, text="biryani", font="times 18")
label12.place(x=20, y=280)
 
e3 = Entry(window)
e3.place(x=20, y=310)
 
label13 = Label(window, text="chicken wings", font="times 18")
label13.place(x=20, y=360)
 
e4 = Entry(window)
e4.place(x=20, y=390)
 
label14 = Label(window, text="tea", font="times 18")
label14.place(x=20, y=420)
 
e5 = Entry(window)
e5.place(x=20, y=450)
 

 # execute calculate function after 1 second
window.after(1000, calculate)
 
# closing the main loop
window.mainloop()


# In[ ]:





# In[ ]:




```
